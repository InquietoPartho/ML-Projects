```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
sonar_df = pd.read_csv('/content/sonar data.csv',header=None)
```

```python
sonar_df.head()
```

```python
sonar_df.shape
```

```python
sonar_df.describe()
```

```python
sonar_df[60].value_counts()
```

```python
sonar_df.groupby(60).mean()
```

```python
#seperating data and label
X = sonar_df.drop(columns=60,axis=0)
Y = sonar_df[60]
```

```python
print(X)
print(Y)
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size = 0.1, stratify = Y, random_state=1)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
print(X_train)
print(Y_train)
```

#Model Trainning --> Logistic Regression

```python
model = LogisticRegression()
```

```python
#Training the logistic regression model
model.fit(X_train,Y_train)
```

#Model Evalution

```python
#accuracy check
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
```

```python
print('Accurary of training data: ', training_data_accuracy)
```

```python
#accuracy check
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction,Y_test)
```

```python
print('Accurary of test data: ', test_data_accuracy)
```
